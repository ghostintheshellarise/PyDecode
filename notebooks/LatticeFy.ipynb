```python
import fst
import numpy as np
import numpy.random
import pydecode.test.utils as utils
```

```python
hypergraph = utils.simple_hypergraph()
```

```python
syms = fst.SymbolTable()
for i in range(len(hypergraph.edges)):
    syms[str(i)] = i + 1
```

```python
a = fst.StdAcceptor(syms)
def up(i): return len(hypergraph.nodes) + i
for node in hypergraph.nodes:
    if len(node.edges):
        for edge in node.edges:
            a.add_arc(edge.head.id, edge.tail[0].id, str(edge.id), 0.0)
            for i in range(len(edge.tail) - 1):
                a.add_arc(up(edge.tail[i].id), edge.tail[i+1].id, fst.EPSILON, 0.0)
            a.add_arc(up(edge.tail[len(edge.tail) - 1].id), up(edge.head.id), str(up(edge.id)), 0.0)
    else:
        a.add_arc(node.id, up(node.id), "N" + str(node.id), 0.0)
a.start = hypergraph.root.id
a[up(hypergraph.root.id)].final = True
```

```python
weights = numpy.random.random(len(hypergraph.edges))
```

```python
for state in a:
    for arc in state.arcs:
        #print arc.ilabel, a.isyms.find(arc.ilabel)
        if arc.ilabel > 0 and arc.ilabel <= len(hypergraph.edges): 
            arc.weight = fst._fst.TropicalWeight(weights[arc.ilabel - 1])
            print weights[arc.ilabel - 1]
```

```python
def relax(D, node):
    for p in D.predecessors(node):
        for s in D.successors(node):
            D.add_edge(p, s)
    D.remove_node(node)
```
