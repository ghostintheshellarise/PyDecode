# Tutorial 2: Edit Distance

```python
import pydecode
import numpy.random
import numpy as np
```

In this tutorial, we will consider solving an edit distance problem using PyDecode. This will provide an overview of 

* Constructing a dynamic program.
* Assigning scores to operations. 
* Using more advanced queries.

For an overview of the edit distance problem, see the wikipedia article on Levenshtein distance, http://en.wikipedia.org/wiki/Levenshtein_distance.


```python
kInsert, kDelete, kMatch = 0, 1, 2
operations = [kInsert, kDelete, kMatch]
t = {kInsert: (1,0), kDelete: (0,1), kMatch: (1,1)}
s = {kInsert: "<", kDelete: ">", kMatch: "="}
```

```python
def edit_distance(str1, str2):
    c = pydecode.ChartBuilder(
        item_set=pydecode.IndexSet((m+2, n+2)),
        output_set=pydecode.IndexSet(3))
    m, n = len(str1), len(str2)
    for i, s_char in enumerate(str1+"_"):
        for j, t_char in enumerate(str2+"_"):
            if (i,j) == (0, 0): 
                c[0, 0] = c.init()
            else:
                c[i, j] = \
                    [c.merge(key, values=[op])
                     for op in operations
                     if op != kMatch or str1[i-1] == str2[j-1]
                     for key in [(i-t[op][0], j-t[op][1])]
                     if key[0] >= 0 and key[1] >= 0]
    return c
```

```python
def show(str1, str2, ops):
    pos = [0, 0]
    for op in ops:
        print "%1s %1s %1s"%((str1+"_")[pos[0]], s[op], (str2+"_")[pos[1]])
        pos[0] += t[op][0]
        pos[1] += t[op][1]
```

```python
str1, str2 = "aab", "bab"
chart = edit_distance(str1, str2)
hypergraph = chart.finish()
matrix = chart.matrix()
item_matrix = chart.item_matrix()

hypergraph.labeling = pydecode.Labeling(hypergraph, 
                                        ["%s | %s"%((str1 + "_")[key[0]], (str2 + "_")[key[1]]) 
                                         for index in chart.item_matrix().nonzero()[0] 
                                         for key in [chart.item_set.element(index)]], 
                                        [s[i] for i in chart.matrix().nonzero()[0]])
    
```

```python
import pydecode.display as display
display.HypergraphFormatter(hypergraph).to_ipython()
```

```python
output_scores = np.zeros(3)
output_scores[kMatch] = 1.0
theta = chart.matrix().T * output_scores
```

```python
path = pydecode.best_path(hypergraph, theta)
display.HypergraphPathFormatter(hypergraph, [path]).to_ipython()
```

```python
str1, str2 = "hell this is a longer sequence", "hello ths is a longr seqence"
chart = edit_distance(str1, str2)
hypergraph = chart.finish()
theta = chart.matrix().T * output_scores
path = pydecode.best_path(hypergraph, theta)
mat = chart.matrix()


show(str1, str2, [mat.T[edge.id].nonzero()[1][0] for edge in path])
```
